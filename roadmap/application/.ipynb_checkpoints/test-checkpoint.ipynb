```python
import requests
import pandas as pd
from pandas import DataFrame,Series
import datetime as dt
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fetch_quandl(ticker, apiKey) :
    ticker = ticker.upper()
    now = dt.datetime.now().date()
    then = now - dt.timedelta(days=300)
    dates = pd.date_range(then,now)
    df1 = pd.DataFrame(index=dates)
    then = "&start_date=" + then.strftime("%Y-%m-%d")
    now  = "&end_date=" + now.strftime("%Y-%m-%d")
    reqUrl = 'https://www.quandl.com/api/v3/datasets/WIKI/' + ticker + \
					'.json?api_key=' + apiKey + now + then
    print reqUrl

    r = requests.get(reqUrl)

    if r.status_code < 400 :
    # get name of company
        name = r.json()['dataset']['name']
	name = name.split('(')[0]
    # get data
	dat = r.json()['dataset']
	df = DataFrame(dat['data'], columns=dat['column_names'] )
	df = df.set_index(pd.DatetimeIndex(df['Date']))
    else :
	print "Stock ticker not valid"
	df = None
	name = None

    return df1.join(df).dropna(), name

def rolling_stats(df):
    period = 10
    df['Rolling_mean'] = pd.rolling_mean(df['Adj. Close'],window=period)
    df['Rolling_std'] = pd.rolling_std(df['Adj. Close'],window=period)
    df['Bollinger_band_low'] = pd.rolling_mean(df['Adj. Close'],window=period) - 2*pd.rolling_std(df['Adj. Close'],window=period)
    df['Bollinger_band_high'] = pd.rolling_mean(df['Adj. Close'],window=period) + 2*pd.rolling_std(df['Adj. Close'],window=period)
    return df,period

```

```python
apiKey = 'KhAguNLDB2kxRQn5u_xi'
ticker = 'GOOG'
df,name = fetch_quandl(ticker, apiKey)
df,period = rolling_stats(df)
```

```python
xtrain = df[['Close', 'Rolling_mean','Rolling_std','Volume']][period:-7]
ytrain = df[['Adj. Close']][period+7:]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize=(12, 10))
ax0, ax1, ax2, ax3 = axes.flat

x = xtrain['Close'].values
y = ytrain.values
fit = np.polyfit(x,y,1)
ax0.scatter(x,y,color='blue',s = 50,alpha=0.5)
ax0.plot(x, fit[0] * x + fit[1], color='red')
ax0.set_xlabel('Close')
ax0.set_ylabel('Adj. Close')

x = xtrain['Rolling_mean'].values
y = ytrain.values
fit = np.polyfit(x,y,1)
ax1.scatter(x,y,color='blue',s = 50,alpha=0.5)
ax1.plot(x, fit[0] * x + fit[1], color='red')
ax1.set_xlabel('Rolling_mean')
ax1.set_ylabel('Adj. Close')

x = xtrain['Rolling_std'].values
y = ytrain.values
fit = np.polyfit(x,y,1)
ax2.scatter(x,y,color='blue',s = 50,alpha=0.5)
ax2.plot(x, fit[0] * x + fit[1], color='red')
ax2.set_xlabel('Rolling_std')
ax2.set_ylabel('Adj. Close')

x = xtrain['Volume'].values
y = ytrain.values
fit = np.polyfit(x,y,1)
ax3.scatter(x,y,color='blue',s = 50,alpha=0.5)
ax3.plot(x, fit[0] * x + fit[1], color='red')
ax3.set_xlabel('Volume')
ax3.set_ylabel('Adj. Close')
```
